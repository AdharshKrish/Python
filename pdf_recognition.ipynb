```python
# Python program to illustrate HoughLine 
# method for line detection 
import cv2 
import numpy as np 

# Reading the required image in 
# which operations are to be done. 
# Make sure that the image is in the same 
# directory in which this python program is 
img = cv2.imread('studentlist1new.jpeg') 

# Convert the img to grayscale 
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) 

# Apply edge detection method on the image 
edges = cv2.Canny(gray,50,150,apertureSize = 3) 

# This returns an array of r and theta values 
lines = cv2.HoughLines(edges,1,np.pi/180, 200) 

# The below for loop runs till r and theta values 
# are in the range of the 2d array 
for r,theta in lines[0]: 
	
	# Stores the value of cos(theta) in a 
	a = np.cos(theta) 

	# Stores the value of sin(theta) in b 
	b = np.sin(theta) 
	
	# x0 stores the value rcos(theta) 
	x0 = a*r 
	
	# y0 stores the value rsin(theta) 
	y0 = b*r 
	
	# x1 stores the rounded off value of (rcos(theta)-1000sin(theta)) 
	x1 = int(x0 + 1000*(-b)) 
	
	# y1 stores the rounded off value of (rsin(theta)+1000cos(theta)) 
	y1 = int(y0 + 1000*(a)) 

	# x2 stores the rounded off value of (rcos(theta)+1000sin(theta)) 
	x2 = int(x0 - 1000*(-b)) 
	
	# y2 stores the rounded off value of (rsin(theta)-1000cos(theta)) 
	y2 = int(y0 - 1000*(a)) 
	
	# cv2.line draws a line in img from the point(x1,y1) to (x2,y2). 
	# (0,0,255) denotes the colour of the line to be 
	#drawn. In this case, it is red. 
	cv2.line(img,(x1,y1), (x2,y2), (0,0,255),20) 
	
# All the changes made in the input image are finally 
# written on a new image houghlines.jpg 
cv2.imwrite('linesDetected.jpg', img) 
```

```python
#logic - 2

from wand.image import Image
import pytesseract
from PIL import Image as oimg
import numpy as np
import cv2
import os
import shutil
import csv

if os.path.exists('image recognition\\'):
    print("deleting image recognition folder...")
    shutil.rmtree("image recognition")
pdfname = "studentlist.pdf"
list=[]
imglist=[]
index=1
print("opening pdf...")
with(Image(filename=pdfname, resolution=480)) as source:
    print("pdf opened")
    images = source.sequence
    pages = len(images)
    os.mkdir("image recognition")
    os.chdir("image recognition")
    for i in range(pages):
        jpegname = 'page' + str(i+1) + '.jpeg'
        print("---> "+jpegname[:-5]+" <---")
        Image(images[i]).save(filename=jpegname)
        imglist.append(cv2.imread(jpegname))
        img = cv2.vconcat(imglist)
        print("replacing "+jpegname)
        cv2.imwrite(jpegname,img)
        edges = cv2.Canny(img, 80, 120)
        lines = cv2.HoughLinesP(image=edges,rho=3,theta=np.pi/180, threshold=3400,lines=np.array([]), minLineLength=3400,maxLineGap=20)
        print("detecting lines...")
        a,b,c=lines.shape
        l=[]
        for i in range(a):
            pt1 = (lines[i][0][0],lines[i][0][1])
            pt2 = (lines[i][0][2],lines[i][0][3])
            if(pt1[0]>230 and pt1[0]<240 and pt2[0]>3738 and pt2[0]<3746):
                l.append(pt1[1])
        l.sort()
        mew=[]
        if(l[0]!=0):
            mew.append(0)
        for i,n in enumerate(l):
            if(i%2==0):
                mew.append(n)
        mew.append(img.shape[0])
        print("lines detected as ",mew)
        with Image(filename=jpegname) as source:
            for line in range(len(mew)-1):
                with source[0:img.shape[1],mew[line]:mew[line+1]] as cropped:
                    index+=1
                    segname='segment'+str(index)+'.jpeg'
                    Image(cropped).save(filename=segname)
                    print("cropped "+segname)
        del imglist
        imglist=[]
        imglist.append(cv2.imread(segname))
        print("deleting "+jpegname+" and "+segname+"...")
        os.remove(segname)
        os.remove(jpegname)
        index-=1
print("done ;-)")
os.chdir("..")
```
